**How many languages do you speak? Let's look at our pool of users**

I know:

**Start by replacing all the empty cells in the 'speaks' column with 'no answer'.  We assume these users speak only 1 language (most likely English) and therefore did not respond to the question.**

```python
df.loc[df.speaks.isna(), 'speaks'] = 'no answer' # With loc there is only one call to __getitem__. #.isna() or .isnull() do the same thing
df.speaks = np.where(df.speaks.isna(), 'no answer', df.speaks)
df.speaks = df.speaks.where(~df.speaks.isna(), 'no answer')
df.speaks = df.speaks.mask(df.speaks.isna(), 'no answer')
df.speaks = df.speaks.apply(lambda x: 'no answer' if pd.isna(x) else x)
df.speaks.fillna("no answer", inplace = True) 
df.speaks.replace(np.nan, "no answer", inplace=True)

#Check the result:
print('The number of empty cells in speak column: {}\n'.format(df.speaks.isnull().sum()))
```

**Now add a column called 'languages' to the dataframe which holds the number of languages each user speaks**

```python
df["languages"] = df["speaks"].apply(lambda x: len(x.split(","))) # or len(x.split(","))+1
df["languages"] = df.speaks.str.count(',') + 1
```

**Finally use hist to show a histogram of the number of users that speak a certain number of languages**

```python
xlabel = 'number of languages a person speaks'

fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=4, figsize=(8,17), constrained_layout=True)
fig.suptitle('Histogram of number of users that speak a certain number of languages')

# For continous variables
sns.histplot(data=df, x="languages", bins=5, ax=ax1, color='purple')
ax1.set(xlabel=xlabel, title='Bins=5', ylabel='Count')

sns.histplot(data=df, x="languages", bins=10, ax=ax2, color='g')
ax2.set(xlabel=xlabel, title='distplot', ylabel='Count')

#For discrete variables
sns.countplot(x='languages',data=df, color="C0", ax=ax3, palette='Reds')
ax3.set(title='Countplot', ylabel='Count')

sns.countplot(x='languages',data=df, color="C0", ax=ax4, palette='plasma')
ax4.set(title='Countplot - The xlabel has changed', xlabel=xlabel, ylabel='Count')

plt.show()
```

**Let's take a closer look at 4 different fields and try to determine if they are male or female dominated**

**The four fields we will look at are:**

**1. Hi-tech jobs which include 'science / tech / engineering' and 'computer / hardware / software'**

**2. Money related jobs which include 'banking / financial / real estate'**

**3. Jobs in the medical field including 'medicine / health'**

**4. Jobs in the arts 'artistic / musical / writer'**

```python
import pandas as pd
import matplotlib.pyplot as plt

map_to_type = {'computer / hardware / software': 'Hi-tech Jobs',
               'science / tech / engineering': 'Hi-tech Jobs',
               'banking / financial / real estate': 'Money Jobs',
               'medicine / health': 'Medical Jobs',
               'artistic / musical / writer': 'Art Jobs',
              }

df['job_type'] = df['job'].map(map_to_type)

df_job_sex = pd.crosstab(index=df['sex'], columns=df['job_type'])

fig, axes = plt.subplots(2, 2, figsize=(12, 6))

df_job_sex.loc[:, ['Hi-tech Jobs']].plot(kind='pie', subplots=True, legend=False, labels=['Female', 'Male'],
                autopct='%.1f%%', colors=['g', 'b'], startangle=90,ylabel='', ax=axes[0][0], labeldistance=1.3, wedgeprops={'linewidth': 0})
axes[0][0].set_title('Hi-tech Jobs')

df_job_sex.loc[:, ['Money Jobs']].plot(kind='pie', subplots=True, legend=False, labels=['Female', 'Male'],
                autopct='%.1f%%', colors=['g', 'b'], startangle=90,ylabel='', wedgeprops={'linewidth': 0},labeldistance=1.3,
                ax=axes[0][1])
axes[0][1].set_title('Money Jobs')

df_job_sex.loc[:, ['Medical Jobs']].plot(kind='pie', subplots=True, legend=False, labels=['Female', 'Male'],
                autopct='%.1f%%', colors=['g', 'b'], startangle=90,ylabel='', wedgeprops={'linewidth': 0},labeldistance=1.3,
                ax=axes[1][0])
axes[1][0].set_title('Medical Jobs')

df_job_sex.loc[:, ['Art Jobs']].plot(kind='pie', subplots=True, legend=False, labels=['Female', 'Male'],
                autopct='%.1f%%', colors=['g', 'b'], startangle=90,ylabel='', wedgeprops={'linewidth': 0},labeldistance=1.3,
                ax=axes[1][1], title='Male/female ratio in Art Jobs')
axes[1][1].set_title('Art Jobs')

for ax in axes.flatten():
    ax.set_aspect('auto')
    ax.invert_xaxis()

plt.show()
```
