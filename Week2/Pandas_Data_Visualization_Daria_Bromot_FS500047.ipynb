** Recreate this scatter plot of b vs a. Note the color and size of the points. Also note the figure size. See if you can figure out how to stretch it in a similar fashion. Remeber back to your matplotlib lecture...**

```python
# Create a scatter plot
fig, ax = plt.subplots(figsize=(12,3))
ax.scatter(df3['a'], df3['b'], c='red', edgecolors='black', s=50)

# Set minimum and maximum values for the X and Y axis
ax.set_xlim(xmin=-0.2, xmax=1.2)
ax.set_ylim(ymin=-0.2, ymax=1.2)

# Display the scatter plot
plt.show()
```

** Create a histogram of the 'a' column.**

```python
# Create a histogram
ax = df3['a'].plot.hist(color='blue', edgecolor='black')

# Set minimum and maximum values for the X and Y axes
ax.set_xlim(xmin=0, xmax=1)
ax.set_ylim(ymin=0, ymax=70)

# Customize the properties of tick labels and tick marks on axes
ax.tick_params(axis='both', which='both', direction='in', length=4, width=0.5, colors='black', top=True, right=True, left=True)

# Display the histogram
plt.show()
```

** Create a boxplot comparing the a and b columns.**

```python
import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax = plt.subplots()

# Create a boxplot with dotted whiskers


ax.boxplot([df3['a'], df3['b']], whis=1.5,
           boxprops=dict(linewidth=1.5, color="#e38678"),
           whiskerprops=dict(linewidth=1.5, linestyle='--', dashes=(6, 3),  color='#e38678'))

ax.set_xticklabels(['a', 'b'])

# Add markup in a circle from the outside
ax.tick_params(axis='both', which='both', direction='out', length=4, width=0.5, colors='gray', top=True, right=True, left=True)

# Set the minimum and maximum values for the axes
ax.set_ylim(ymin=0, ymax=1)

# Display the graph
plt.show()
```

** Create a kde plot of the 'd' column **

```python
fig, ax = plt.subplots()

ax=df3['d'].plot.kde()

# Set minimum and maximum values for the X and Y axes
ax.set_xlim(xmin=-0.5, xmax=1.5)
ax.set_ylim(ymin=0, ymax=1.2)

# Customize the properties of tick labels and tick marks on axes
ax.tick_params(axis='both', which='both', direction='out', length=4, width=0.5, colors='gray', top=True, right=True, left=True, bottom=True)

# set the x ticks
ax.set_xticks(np.arange(-0.5, 1.6, 0.5))

# Display the histogram
plt.show()
```

** Create an area plot of all the columns for just the rows up to 30. (hint: use .ix).**

```python
ax=df3.loc[0:30].plot.area(alpha=0.4)

# Set minimum and maximum values for the X and Y axes
ax.set_xlim(xmin=0, xmax=30)
ax.set_ylim(ymin=0, ymax=3)

# Customize the properties of tick labels and tick marks on axes
ax.tick_params(axis='both', which='both', direction='out', length=4, width=0.5, colors='gray', top=True, right=True, left=True, bottom=True)

# Create the legend with a fontsize of 12 and 2 columns
legend = ax.legend(fontsize=12.5)

# Set the alpha value of the legend to 1.0 to make it not transparent
legend.get_frame().set_alpha(1.0)

# Set the color of the legend border to white
legend.get_frame().set_edgecolor('white')
```

** Notice how the legend in our previous figure overlapped some of actual diagram. Can you figure out how to display the legend outside of the plot as shown below?**

** Try searching Google for a good stackoverflow link on this topic. If you can't find it on your own - [use this one for a hint.](http://stackoverflow.com/questions/23556153/how-to-put-legend-outside-the-plot-with-pandas)**

```python
ax=df3.loc[0:30].plot.area(alpha=0.4)

# Set minimum and maximum values for the X and Y axes
ax.set_xlim(xmin=0, xmax=30)
ax.set_ylim(ymin=0, ymax=3)

# Customize the properties of tick labels and tick marks on axes
ax.tick_params(axis='both', which='both', direction='out', length=4, width=0.5, colors='gray', top=True, right=True, left=True, bottom=True)

# Create the legend with a fontsize of 12
legend_lines = [plt.Line2D([0], [0], label="a", linestyle='-', linewidth=0.5, color='#d57a6f'),
                plt.Line2D([0], [0], label="b", linestyle='-', linewidth=0.5, color='#789fcb'),
                plt.Line2D([0], [0], label="c", linestyle='-', linewidth=0.5, color='#9e98bd'),
                plt.Line2D([0], [0], label="d", linestyle='-', linewidth=0.5, color='#9e9e9e')]

legend = ax.legend( fontsize=12.5, loc='center left',  bbox_to_anchor=(1, 0.5), handles=legend_lines, handlelength=1.5)

# Set the alpha value of the legend to 1.0 to make it not transparent
legend.get_frame().set_alpha(1.0)

# Set the color of the legend border to white
legend.get_frame().set_edgecolor('white')
```
